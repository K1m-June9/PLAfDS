### 선형 가중 결합
- 여러 변수마다 가중치를 다르게 주어 정보를 혼합하는 방법
    - 선형 혼합 또는 가중 결합이라고도 함(선형이라고 가정)

$
W = \lambda_1 v_1 + \lambda_2 v_2 + ... + \lambda_n v_n
$

- 여기서 모든 벡터 $v_i$의 차원은 같다고 가정(안그러면 못더함)
- $\lambda$는 0을 포함한 임의의 실수가 될 수 있음

```python
# 선형 가중 결합
l1 = 1
l2 = 2
l3 = 3
v1 = np.array([4,5,1])
v2 = np.array([-4,0,-4])
v3 = np.array([1,3,2])

res = l1*v1 + l2*v2 + l3*v3
print(res)
```
